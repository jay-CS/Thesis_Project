```python
import glob
import numpy as np

lst_0 = [0.0]*11
row_0 = [0.0]*64

for k_num in ['50','75','100']:
    path = r"/home/jupyter/src/Thesis_Project/Data/"+k_num
    all_files = glob.glob(path + "/*")
    all_files.sort()
    for file in all_files:

        keystrokes = list(np.load(file))

        # Pad the array to shape (68, 5, 64, 64)
        keystrokes = np.pad(keystrokes, ((0,0), (0,0), (11,11), (11,11)), mode='constant')

        # Verify the shape of the padded array
        print(file.split('.')[-2][-3:],keystrokes.shape)
        np.save(r"/home/jupyter/src/Thesis_Project/Data_padded/"+k_num+"/" + file.split('.')[-2][-3:]+".npy",keystrokes)
```
